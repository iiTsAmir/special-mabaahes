```python
from IPython.display import clear_output as co

print('(ﾉ ﾟｰﾟ)ﾉ WELCOME TO 2 PLAYER WORD GUESSING GAME\n')
input('press enter to start...')
Round = 0
ply = True
dif = '5'                                               # dif = difficulty

co()                                                    # clear output
wp = (input('enter first player name: '))               # wp = word player
gp = (input('enter second player name: '))              # gp = guess player

while ply == True:
    Round += 1
    
    co()
    print(f'round {Round} • - - - - - - - - - - - - - - - - - - -')
    print(f'   {wp}  ==>  *_*     {gp}  ==>  >_<\n')

    word = [*input(f'{wp} enter your word: ')]          # word save the wp word
    gword =[]                                           # gword hold gp progress
    
    answer=''                                           # answer hold right
    for i in word:                                       # word and shows
        answer+=i                                         # when round finish
    
    won = False                                         # won use to check which player won
    true = False                                        # true use to check gess is true or false :)
    chances = int(dif)
    
    for i in word:                                      # create gword with word
        if i == ' ':
            gword.append(' ')
        else:
            gword.append('_')
    co()
    print(f'round {Round} • - - - - - - - - - - - - - - - - - - -')
    print(f'{gp} start guessing...')

    out = ''                                            # out print gword for first time
    for i in gword:
        out = out + i + ' '
    print(out[:-1]+'\n')
    out=''
    
    while chances > 0:
        
        if   int(chances) == int(dif) and true == False :
            gus = input(f'you have {chances} chances, enter a character to start ')
        elif chances > 2:
            gus = input(f'you have {chances} chances, enter next character ')
        elif chances > 1:
            gus = input(f'you only have {chances} chances, be careful!  ')
        else:
            gus = input(f'its yor last chance, i think {wp} gonna win  (〜￣▽￣)〜  ')

        if len(gus)>1:
            print('enter a single character!')
            continue

        true = False
        
        for i in range (len(word)):                     # check the geuss
            if gus == word[i]:
                gword[i]=word[i]
                true = True
        
        if true == False:                               # decrease chachances
            chances -= 1
        
        for i in gword:                                 # out print gword after every guess
            out = out + i + ' '
        print(out[:-1]+'\n')
        out=''
        
        if gword == word:                               # check who is winner
            won = True
            break
    if won:
        print(f'{gp} is winner!!')
        print(f'the word was "{answer}"\n')
    else:
        print(f'{wp} is winner!!')
        print(f'the word was "{answer}"\n')

    
    pn = ''                                             #pn = play next round
    j = 1
    
    while pn != 'N' and pn != 'F' and pn != 'D':        # ask for play next round
        print('N'*j + ' to next round')
        print('D'*j + ' to change difficulty')
        print('F'*j + ' to finish the game\n')
        pn = input().upper()
        
        j += 1
        if j == 5:                                      # check players mental health :)
            co()
            print('(╯▔皿▔)╯  moshkel daariii??!!!!')
            ply = False
            break
    wp , gp = gp , wp
    if pn == 'D':                                      # set difficulty
        co()
        ok = True
        while ok:
            dif = input('enter number of chances(1-15): ')
            if dif.isdigit():
                if int(dif) <= 15 and int(dif) >= 1:
                    chances = dif
                    break
                else:
                    print('between 1 and 15 !!\n')
            else:
                print('enter number !!\n')
     
    if pn == 'F':                                       # check play next round
        co()
        print('good bye')
        ply = False
```
